```python
#api call to get roster info
seasons= list(range(20142015, 20222023, 10001))
rosters = []
url= "https://statsapi.web.nhl.com/api/v1/teams/13?expand=team.roster&season="
for season in seasons:
    link = f"{url}{season}"
    r = requests.get(link, params={"Content-Type": "application/json"})
    data = r.json()

    df = pd.DataFrame(data['teams'][0]['roster']['roster'])
    person = df.person
    persondf = pd.json_normalize(person)
    persondf['Season']= pd.Series([format(season) for x in range(len(persondf.index))])
    rosters.append(persondf)
    
combined_df = pd.concat([roster for roster in rosters], ignore_index=True, axis=0)
combined_df
```

```python
#list(range(20142015, 20222023, 10001))
```

```python
#api call to get player descriptive info [hieght, age, weight]
players = combined_df['id'].tolist()
#players_url = "https://statsapi.web.nhl.com/api/v1/people/"
stats = []
for player in players:
    link = f"https://statsapi.web.nhl.com/api/v1/people/{player}"
    r = requests.get(link, params={"Content-Type": "application/json"})
    data = r.json()
    
    df = pd.json_normalize(data['people'])
    stats.append(df)
    
combined_stats_df = pd.concat([stat for stat in stats], ignore_index=True, axis=0)
combined_stats_df
```

```python
#combined_df = pd.DataFrame(combined_df)
combined_dfx = []
combined_dfx= pd.DataFrame()
combined_dfx['ID'] = combined_df['id']
combined_dfx['Season'] = combined_df['Season']
combined_dfx
```

```python
#age height and weight data

combined_stats_dfx = pd.DataFrame()
combined_stats_dfx['id'] = combined_stats_df['id']
combined_stats_dfx['birthDate'] = combined_stats_df['birthDate']
combined_stats_dfx['Height'] = combined_stats_df['height']
combined_stats_dfx['Weight'] = combined_stats_df['weight']
combined_stats_dfx
```

```python
merge_player_info = pd.concat([combined_dfx, combined_stats_dfx.reindex(combined_stats_dfx.index)], axis=1)
merge_player_info
```

```python
# Age calculation 
merge_player_info['Seasonx'] = merge_player_info['Season'].str[4:]

merge_player_info['Seasonx'] = merge_player_info['Seasonx'].astype(str) + '-01-31'

merge_player_info['Seasonx'] = merge_player_info['Seasonx'].apply(pd.to_datetime)
merge_player_info['birthDate'] = merge_player_info['birthDate'].apply(pd.to_datetime)

merge_player_info['Age'] = (merge_player_info['Seasonx']) - (merge_player_info['birthDate'])
merge_player_info['Age'] = merge_player_info['Age'] / np.timedelta64(1, 'Y')

merge_player_info = merge_player_info.drop(['birthDate','Seasonx','id'], axis=1)

merge_player_info = merge_player_info[['ID','Season','Age','Height','Weight']]

merge_player_info
```

```python
#api call to get player stats for each season
season_list = combined_df['Season'].tolist() #x
player_list = combined_df['id'].tolist() #y
infos = []
for x, y, in zip(season_list, player_list):
    link = f"https://statsapi.web.nhl.com/api/v1/people/{y}/stats?stats=statsSingleSeason&season={x}"
    r = requests.get(link, params={"Content-Type": "application/json"})
    data = r.json()
    
    df= pd.json_normalize(data['stats'][0]['splits'])
    
    df['ID'] = y
    infos.append(df)
    combined_des_stats_df = pd.concat([info for info in infos], ignore_index=True, axis=0)
combined_des_stats_df
```

```python
combined_des_stats_df.columns = combined_des_stats_df.columns.str.replace('^stat.', '')
combined_des_stats_dfs = combined_des_stats_df.fillna(0)
combined_des_stats_dfs
```

```python
df1 = []
df2= []
df1 = combined_des_stats_dfs.reset_index(drop=True)
df2 = merge_player_info.reset_index(drop=True)

panthers_players_table = pd.merge(df2, df1, how='inner', 
                  left_on=['ID', 'Season'],
                  right_on=['ID', 'season'])

panthers_players_table

panthers_players_table = panthers_players_table.drop(['season'], axis=1)

panthers_players_table
```

```python
# api call for boxscore Indivuital game ID from here
seasons= list(range(20142015, 20222023, 10001))

gameIDs = []
for season in seasons:
    statlink = f"https://statsapi.web.nhl.com/api/v1/schedule?teamId=13&season={season}&gameType=R"
    r = requests.get(statlink, params={'Content-Typer': "application/json"})
    data = r.json()
    dates = range(len(data['dates']))
    for date in dates:
        gameIDs.append(data['dates'][date]['games'][0]['gamePk'])
    
```

```python
#games when Panthers are away
gameIDs = gameIDs
game_a_stats = []
game_stats_a_df = []
i = 0
d=[]
e=[]
x=[]
for gameID in gameIDs:
    link = f"https://statsapi.web.nhl.com/api/v1/game/{gameID}/boxscore"
    r = requests.get(link, params={'Content-Typer': "application/json"})
    data = r.json()
    tID = data['teams']['away']['team']['id']

    if tID == 13:
        #i = i+1
        #print('pantenrs are away',i)
        d = pd.json_normalize(data['teams']['away']['teamStats']['teamSkaterStats'])
        d['gameID'] = gameID
        d['Home/Away'] = 'Away'
        #print(gameID)
        e = pd.json_normalize(data['teams']['home']['teamStats']['teamSkaterStats'])
        e['gameID'] = gameID
        e = e.add_prefix('OPP')
        homeID = data['teams']['home']['team']['id']
        #print(homeID)
        
        x = pd.concat([d, e.reindex(e.index)], axis=1)
        game_a_stats.append(x)
    
    else:
        #i= i+1
        pass
    
game_stats_a_df = pd.concat([game_a_stat for game_a_stat in game_a_stats], ignore_index=True, axis=0)
```

```python
#Games when Panthers are home
gameIDs = gameIDs
game_h_stats = []
game_stats_h_df = []
f=[]
g=[]
y=[]
i = 0
for gameID in gameIDs:
    link = f"https://statsapi.web.nhl.com/api/v1/game/{gameID}/boxscore"
    r = requests.get(link, params={'Content-Typer': "application/json"})
    data = r.json()
    
    if (data['teams']['home']['team']['id']) == 13:
        #i=i+1
        #print('Alexa Play Bennie and the Jetse, Panthers are home', i, gameID)
        f= pd.json_normalize(data['teams']['home']['teamStats']['teamSkaterStats'])
        f['gameID'] = gameID
        f['Home/Away'] = 'Home'
        g = pd.json_normalize(data['teams']['away']['teamStats']['teamSkaterStats'])
        g['gameID'] = gameID
        g = g.add_prefix('OPP')
        
        y = pd.concat([f, g.reindex(g.index)], axis=1)
        game_h_stats.append(y)
    
    else:
        pass
    

game_stats_h_df = pd.concat([game_h_stat for game_h_stat in game_h_stats], ignore_index=True, axis=0)
```

```python
#API call for getting the last period of the games
gameIDs = gameIDs
last_period_dfs = []
lp=[]

for gameID in gameIDs:
    
    statlink = f"https://statsapi.web.nhl.com/api/v1/game/{gameID}/linescore"
    r = requests.get(statlink, params={'Content-Typer': "application/json"})
    data = r.json()
    
    print('GameID:',gameID)
    lp = data['periods'][-1]['periodType']
    last_period_dfs.append(lp)
    
    
last_period_dfs = pd.DataFrame(last_period_dfs)
last_period_dfs = last_period_dfs.rename(columns={0:'last'})
last_period_dfs
```

```python
#Merging home and away game stats
results = []
frames = []
frames = [game_stats_h_df, game_stats_a_df]
results =pd.concat(frames)

panthers_boxscores = results.sort_values(by=['gameID'])

panthers_boxscores = panthers_boxscores.reset_index(drop=True)
panthers_boxscores

#panthers_boxscoresx = pd.concat([panthers_boxscores, last_period_dfs], axis=1)
panthers_boxscores['last'] = last_period_dfs['last']
panthers_boxscores

#contitional for creating outcome column
conditions = [
    (panthers_boxscores['goals'] > panthers_boxscores['OPPgoals']),
    ((panthers_boxscores['goals'] < panthers_boxscores['OPPgoals']) & (panthers_boxscores['last'] =='OVERTIME')),   
    (panthers_boxscores['goals'] < panthers_boxscores['OPPgoals']),
    (panthers_boxscores['goals'] == panthers_boxscores['OPPgoals'])]
    
values = ['Win', 'Overtime Loss',' loss', '']
panthers_boxscores['Outcome'] = np.select(conditions, values)
panthers_boxscores

#panthers_boxscores.columns

#rearranged column list
column_list = ['gameID', 'Home/Away','Outcome', 'goals', 'pim', 'shots', 'powerPlayPercentage', 'powerPlayGoals',
       'powerPlayOpportunities', 'faceOffWinPercentage', 'blocked',
       'takeaways', 'giveaways', 'hits',  'OPPgoals',
       'OPPpim', 'OPPshots', 'OPPpowerPlayPercentage', 'OPPpowerPlayGoals',
       'OPPpowerPlayOpportunities', 'OPPfaceOffWinPercentage', 'OPPblocked',
       'OPPtakeaways', 'OPPgiveaways', 'OPPhits']

panthers_game_boxscores_table = panthers_boxscores[column_list]
panthers_game_boxscores_table

panthers_game_boxscores_table.to_csv('Panthers_game_boxscores_table.csv')
```
